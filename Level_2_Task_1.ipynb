# Task: Restaurant Ratings
Analyze the distribution of aggregate
ratings and determine the most common
rating range.
Calculate the average number of votes
received by restaurants.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Dataset.csv")
df_cleaned=df.dropna(subset=['Votes','Aggregate rating'])
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df_cleaned['rating range'] = pd.cut(df_cleaned['Aggregate rating'], bins=bins, labels=labels, include_lowest=True)
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned['Aggregate rating'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()

rating_range_counts = df_cleaned['rating range'].value_counts()
most_common_rating_range = rating_range_counts.idxmax()
print(f" The Restaurant with the most common rating is within the range of {most_common_rating_range}")
```

```python
#Average vote by received by restaurant
average_votes = df_cleaned['Votes'].mean().round(2)
print(f"The average votes received by each restaurants is provided below:\n{average_votes}")
```
