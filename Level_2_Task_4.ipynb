# Task: Restaurant Chains
Identify if there are any restaurant chains
present in the dataset.
Analyze the ratings and popularity of
different restaurant chains.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Dataset.csv")
```

```python
chain_counts = df['Restaurant Name'].value_counts()
restaurant_chains = chain_counts[chain_counts > 1]
restaurant_chains
```

```python
print(f"Number of restaurant chains: {len(restaurant_chains)}")
```

```python
chain_df = df[df['Restaurant Name'].isin(restaurant_chains.index)]
chain_analysis = chain_df.groupby('Restaurant Name').agg(
    average_rating=('Aggregate rating', 'mean'),
    total_votes=('Votes', 'sum'),
    number_of_outlets=('Restaurant Name', 'size')
).sort_values(by='average_rating', ascending=False)
top_chains = chain_analysis.head(10)
top_chains
```

```python
#Visualizing with bar chart
# Bar Chart: Average Ratings
plt.figure(figsize=(10, 6))
top_chains['average_rating'].plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Restaurant Chains by Average Rating')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
print("______________________________________________________________________________________________________________________\n")
# Bar Chart: Total Votes (Popularity)
plt.figure(figsize=(10, 6))
top_chains['total_votes'].plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Restaurant Chains by Popularity')
plt.xlabel('Restaurant Chain')
plt.ylabel('Total Votes')
plt.xticks(rotation=45, ha='right')
plt.show()
```
